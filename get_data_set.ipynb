```python
import os
import subprocess
```

```python
import get_data_set
```

```python
def run_bash_command(bash_command):
    print(bash_command)
    process = subprocess.Popen(bash_command.split(), stdout=subprocess.PIPE)
    output, error = process.communicate()
    return output
```

```python
dir_path = r"/home/erez/github_repos/siva/latest/"
get_data_set.unpack_siva_files_and_checkout_git_repos(dir_path)
```

```python
dir_list = os.listdir(dir_path)
for dir_name in dir_list:
    full_dir_path = os.path.join(dir_path, dir_name)
    dir_files_list = os.listdir(full_dir_path)
    print(full_dir_path, dir_files_list)
    for file_name in dir_files_list:
        file_full_path = os.path.join(full_dir_path, file_name)
        if file_name.endswith(".siva") and os.path.isfile(file_full_path):            
            git_bare_repo_path = os.path.join(full_dir_path, "temp_git_bare_repo_{}").format(file_name)
            os.makedirs(git_bare_repo_path)  
            siva_bash_command = "siva unpack {} {}".format(file_full_path, git_bare_repo_path)
            run_bash_command(siva_bash_command)
            
            git_repo_path = os.path.join(full_dir_path, "temp_git_repo_{}").format(file_name)
            os.makedirs(git_repo_path) 
            git_clone_command = "git clone {} {}".format(git_bare_repo_path, git_repo_path)
            run_bash_command(git_clone_command)
            
            os.chdir(git_repo_path)
            git_branch_command = r"git branch -a"
            output = run_bash_command(git_branch_command)
            print(output)
            
            git_checkout_command = "git checkout {}".format(output.split(b"\n")[0].strip().decode("utf-8"))
            run_bash_command(git_checkout_command)
```
